```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# LOAD DATA
# -------------------------------
df = pd.read_csv("data.csv")

print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

# -------------------------------
# BASIC ANALYSIS
# -------------------------------

# Example: Calculate average of a column (replace 'column_name')
column = 'column_name'

if column in df.columns:
    avg_value = df[column].mean()
    print(f"\nAverage of {column}: {avg_value}")
else:
    print(f"Column '{column}' not found.")

# -------------------------------
# HANDLE MISSING VALUES
# -------------------------------
df = df.fillna(df.mean(numeric_only=True))

# -------------------------------
# BAR CHART
# -------------------------------
# Example: categorical column vs numeric mean
category_col = 'category_column'
value_col = column

if category_col in df.columns:
    grouped = df.groupby(category_col)[value_col].mean()

    grouped.plot(kind='bar')
    plt.title("Bar Chart")
    plt.xlabel(category_col)
    plt.ylabel(f"Average {value_col}")
    plt.show()

# -------------------------------
# SCATTER PLOT
# -------------------------------
# Example: relationship between two numeric columns
x_col = 'column1'
y_col = 'column2'

if x_col in df.columns and y_col in df.columns:
    plt.scatter(df[x_col], df[y_col])
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title("Scatter Plot")
    plt.show()

# -------------------------------
# HEATMAP (CORRELATION MATRIX)
# -------------------------------
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
